```python
df1['sentiment'].replace({'positive':1,'negative':0},inplace=True)
```

```python
df1.head()
```

```python
def clean_htmal(text):
    import re
    clean=re.compile('<.*?>')
    return re.sub(clean,'',text)
```

```python
df1['review']=df1['review'].apply(clean_htmal)
```

```python
df1.head()
```

```python
def convert_lower(text):
    return text.lower()
```

```python
df1['review']=df1['review'].apply(convert_lower)
```

```python
df1.head()
```

```python
def remove_special_character(text):
    x=''
    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x=x+' '
    return x
```

```python
df1['review']=df1['review'].apply(remove_special_character)
```

```python
df1.head()
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
l=0
for i in stopwords.words('english'):
    l=l+1
l
```

```python
def remove_stopword(text):
    x=[]
    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    y=x[:]
    x.clear()
    return y
```

```python
df1['review']=df1['review'].apply(remove_stopword)
```

```python
df1.head()
```

```python
# perform stemming
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem_words(text):
    y=[]
    for i in text:
        y.append(ps.stem(i))
    z=y[:]
    y.clear() 
    return z   
```

```python
df1['review']=df1['review'].apply(stem_words)
```

```python
df1
```

```python
def join_back(list):
    return " ".join(list)
```

```python
df1['review']=df1['review'].apply(join_back)
```

```python
df1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=3500)                # max is 35000
```

```python
X=cv.fit_transform(df1['review']).toarray()
```

```python
X.shape
```

```python
y=df1.iloc[:,-1].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
clf1=GaussianNB()
clf2=MultinomialNB()
clf3=BernoulliNB()
```

```python
clf1.fit(X_train,y_train)
clf2.fit(X_train,y_train)
clf3.fit(X_train,y_train)
```

```python
y_pred1=clf1.predict(X_test)
y_pred2=clf2.predict(X_test)
y_pred3=clf3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('GaussianNB',accuracy_score(y_test,y_pred1))
print('MultinomialNB',accuracy_score(y_test,y_pred2))
print('BernoulliNB',accuracy_score(y_test,y_pred3))
```
